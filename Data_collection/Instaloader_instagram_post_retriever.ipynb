```python
import os
def delete_file_with_extension(dir_name,save_extensions=list()):
    """
        This is a simple file delete function which take extension name as parameter and delete all excluded extension from the directory.
  
        Parameters:
           dir_name (str): Directory name from which file have to be deleted
           save_extensions (list):  List of file extension which don't have to be deleted
 
           
        Return:
            It delete all file with extensions that are not include in the save_extensions parameter.
        """

    directory = os.listdir(dir_name)

    for item in directory:
        for extension in save_extensions:
            if item.endswith(extension):
                pass
            else:
                os.remove(os.path.join(dir_name, item))
```
